### Start time column analysis

```python
us_acc_df.Start_Time = pd.to_datetime(us_acc_df.Start_Time)
```

```python
us_acc_df.Start_Time[0]
```

```python
st_acci = us_acc_df['Start_Time'].unique()
len(st_acci)
```

```python
no_of_start_time_accidents = us_acc_df['Start_Time'].value_counts(ascending=False)
no_of_start_time_accidents
```

```python
no_of_start_time_accidents[:15]
```

```python
sns.distplot(us_acc_df.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True)
```

Insights:
 
  - a high precentage of accidents happens between the time 2pm-6pm
  - second highest precentage of accidents occurs between 6am-9am(most probably people are in hurry to work)

```python
sns.distplot(us_acc_df.Start_Time.dt.dayofweek, bins=7, kde=False, norm_hist=True)
```

Insights:
  
  - lower no of accidents reported on the weekends

Question- is the distribution of accidents by hour the same on weekends as on weekdays? 

```python
sundays_start_time = us_acc_df.Start_Time[us_acc_df.Start_Time.dt.dayofweek == 6]
sns.distplot(sundays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

```python
monday_start_time = us_acc_df.Start_Time[us_acc_df.Start_Time.dt.dayofweek == 0]
sns.distplot(monday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

Insights:

  - On sundays the most occurences of accidents starts from 12am midnight and gradually decreasing till 9am and then again starts increasing gradually unlike weekdays.

```python
sns.distplot(us_acc_df.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```
